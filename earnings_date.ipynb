```python
df = pd.read_csv('Input/tokens_francisco.csv', sep=';')
df
```

```python
tokens =[]
tokens = df['token_id'].tolist()
print(tokens)
print('Hay un total de:' ,len(tokens),'activos')
```

```python
#tokens = ['AAPL', 'BA', 'TSLA', 'KO']  # Example tokens
wait_time = 30

ticker_details = []
# Iterate over each token and fetch financial data
for ticker_id in tokens:
    try:
        # Fetch financial data for each ticker
        details = client.get_ticker_details(ticker=ticker_id)
        ticker_details.append(details)
    except Exception as e:
        print(f"Failed to fetch data for {ticker_id}: {e}")
    time.sleep(wait_time)
#print(ticker_details)
```

```python
# Extracting 'name' and 'ticker' and creating a list of dictionaries
data_for_df = [{'name': detail.name, 'ticker': detail.ticker} for detail in ticker_details]

df = pd.DataFrame(data_for_df)
df
```

```python
df['name_upper'] = df['name'].str.upper()
df['name_upper'] = df['name_upper'].str[:5]
df
```

```python
#tokens = ['AAPL', 'BA', 'TSLA', 'KO']  # Example tokens
wait_time = 30

financial_data = []
# Iterate over each token and fetch financial data
for ticker_id in tokens:
    try:
        # Fetch financial data for each ticker
        for t in client.vx.list_stock_financials(ticker=ticker_id, filing_date_gte='2022-04-19'):
            financial_data.append(t)
    except Exception as e:
        print(f"Failed to fetch data for {ticker_id}: {e}")
    time.sleep(wait_time)

# Optionally, print out the collected financial data to verify
#print(financial_data)
```

```python
financial_data_df = pd.DataFrame(financial_data)
financial_data_df
```

```python
# Convert the 'filing_date' column to datetime
financial_data_df['filing_date'] = pd.to_datetime(financial_data_df['filing_date'])
financial_data_df['earnings'] = financial_data_df['filing_date'] - pd.Timedelta(days=1)
financial_data_df['ticker_id'] = ticker_id

column_order = [
    'ticker_id', 'company_name', 'earnings', 'fiscal_period', 'fiscal_year', 'filing_date'
]

financial_data_df = financial_data_df[financial_data_df['fiscal_period'] != 'TTM']

financial_data_df = financial_data_df[column_order]
financial_data_df
```
