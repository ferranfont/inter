```python
tickers_db = pd.read_csv('tickers_db.csv')
tickers_db = tickers_db[~tickers_db['ticker'].str.endswith('.B')]



# Display the first few rows of the DataFrame
tickers_db.head()
```

```python
tickers_list = tickers_db['ticker'].tolist()

tickers_list
```

```python
import yfinance as yf
import pandas as pd

def get_earnings_dates(ticker_symbol):
    # Create a Ticker object
    ticker = yf.Ticker(ticker_symbol)

    # Get earnings dates and details
    earnings_dates = ticker.earnings_dates

    # Add a column for the ticker symbol
    earnings_dates['Ticker'] = ticker_symbol

    return earnings_dates
```

```python
# Initialize an empty DataFrame
all_earnings = pd.DataFrame()

# Loop through each ticker in the list and concatenate the results
for ticker in tickers_list:
    print(ticker)
    earnings_dates = get_earnings_dates(ticker)
    all_earnings = pd.concat([all_earnings, earnings_dates])
    print(all_earnings)
    time.sleep(10)

# Remove rows where the ticker symbol is 'BF.B'
all_earnings = all_earnings[all_earnings['Ticker'] != 'BF.B']
all_earnings = all_earnings[all_earnings['Ticker'] != 'BRK.B']

# Display the concatenated DataFrame
print("All Earnings Dates and Details:")
print(all_earnings)
```

```python
# Save the DataFrame to a CSV file
all_earnings.to_csv('earnings_calendar.csv', index=False)
```
