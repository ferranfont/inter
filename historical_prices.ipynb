```python
all_market_data = []
for token in tokens:
    getMarketData = ivol.setMethod('/equities/eod/stock-prices')
    marketData = getMarketData(symbol=token)
    # Convert each DataFrame row to a dictionary and flatten it if necessary
    for index, row in marketData.iterrows():
        all_market_data.append(row.to_dict())

# Now create the DataFrame from the list of dictionaries
try:
    top_instruments = pd.DataFrame(all_market_data)
    print(top_instruments)
except Exception as e:
    print(f"Failed to create DataFrame: {e}")
```

```python
top_tokens_historical_prices = pd.DataFrame(all_market_data)
column_renaming = {
    'symbol': 'ticker'
}
top_tokens_historical_prices.rename(columns=column_renaming, inplace=True)
top_tokens_historical_prices
```

```python
# Save the DataFrame to a CSV file with a semicolon delimiter
top_tokens_historical_prices.to_csv('Output/top_tokens_historical_prices.csv', index=False)
```

```python
# plot a sub_dataframe

valor = 'BA'
sub_token = top_tokens_historical_prices[top_tokens_historical_prices['ticker'] == valor]
sub_token['date'] = pd.to_datetime(sub_token['date'])

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(sub_token['date'], sub_token['close'], label='Close Price')

# Adding labels and title
plt.xlabel('Timestamp')
plt.ylabel('Close Price')
plt.title(f'Stock Close Price Over Time: {valor}')

# Rotating x-axis labels for better visibility
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
