```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Function to scrape S&P 500 tickers from Wikipedia
def get_sp500_tickers():
    url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.find('table', {'id': 'constituents'})
    
    tickers = []
    for row in table.findAll('tr')[1:]:
        ticker = row.findAll('td')[0].text.strip()
        tickers.append(ticker)
    
    return tickers

# Get the list of S&P 500 tickers
sp500_tickers = get_sp500_tickers()

# Print the list of tickers
sp500_tickers
```

```python
# Fetch market cap data from Yahoo Finance for each ticker
market_caps = []
for ticker in sp500_tickers:
    try:
        stock = yf.Ticker(ticker)
        info = stock.info
        market_caps.append({
            'Ticker': ticker,
            'Company': info.get('longName', ticker),
            'Market_Cap': info.get('marketCap', 0)
        })
    except KeyError:
        print(f"Data for ticker {ticker} not found. Skipping.")
    except Exception as e:
        print(f"An error occurred for ticker {ticker}: {e}. Skipping.")
```

```python
df_capital = pd.DataFrame(market_caps)
df_capital = df_capital.sort_values(by='Market_Cap', ascending=False)
total_market_cap = df_capital['Market_Cap'].sum()
df_capital['Percentage'] = (df_capital['Market_Cap'] / total_market_cap) * 100
df_capital['Percentage'] =  round(df_capital['Percentage'],2)
df_capital.rename(columns={'Ticker': 'ticker'}, inplace=True)
df_capital.head(20)
```

```python
df_all_market_earnings = pd.read_csv('all_market_earnings.csv')
df_all_market_earnings
```

```python
df_all_market_earnings['earnings_date'] = pd.to_datetime(df_all_market_earnings['earnings_date'])
df_all_market_earnings.dtypes
```

```python
# Create the 'date' column based on the 'earnings_date' column
df_all_market_earnings['date'] = df_all_market_earnings['earnings_date'].apply(
    lambda x: x.date() - pd.Timedelta(days=1) if x.time() <= pd.Timestamp('13:00:00').time() else x.date()
)
#df_all_market_earnings['date'] = df_all_market_earnings['earnings_date'].dt.date
```

```python
df_all_market_earnings.head()
```

```python
# Group by 'date' and aggregate tickers and count the number of rows for each date
date_counts = df_all_market_earnings.groupby('date').agg(
    counts=('ticker', 'size'),
    tickers=('ticker', lambda x: ', '.join(x))
).reset_index()

# Sort the DataFrame by 'counts' in descending order
date_counts = date_counts.sort_values(by='counts', ascending=False).reset_index(drop=True)
```

```python
# Ensure 'ticker' column exists in both DataFrames
if 'ticker' in df_all_market_earnings.columns and 'ticker' in df_capital.columns:
    df_all_market_earnings = df_all_market_earnings.merge(df_capital[['ticker', 'Percentage']], on='ticker', how='left')
else:
    print("One or both DataFrames do not have the 'ticker' column")
```

```python
#df_all_market_earnings.drop(columns=['Percentage_y'], inplace=True)
df_all_market_earnings.drop(columns=['id'], inplace=True)
df_all_market_earnings.rename(columns={'Percentage_x': 'Percentage'}, inplace=True)
```

```python
df = df_all_market_earnings.copy()
df.head(30)
```

```python
# Group by 'date' and perform the required aggregations
date_counts = df.groupby('date').agg(
    players=('ticker', lambda x: list(x)),          # Aggregate tickers into a list
    counts=('ticker', 'size'),                      # Count the number of rows for each date
    impact=('Percentage', 'sum')                    # Calculate the sum of 'Percentage' for each date
).reset_index()
```

```python
date_counts = date_counts.sort_values(by='impact', ascending=False).reset_index(drop=True)

# Display the resulting DataFrame
date_counts.head(30)
```
