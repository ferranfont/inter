```python
import pandas as pd
from datetime import datetime, timedelta

def get_earnings_dates(ticker):
    """
    Get the earnings dates for a specified stock ticker.

    Parameters:
    ticker (str): The stock ticker symbol.

    Returns:
    pandas.DataFrame: A DataFrame containing the earnings dates, or None if there's an error.
    """
    try:
        # Create a Ticker object
        stock = yf.Ticker(ticker)

        # Get the earnings dates
        earnings = stock.earnings_dates.reset_index()

        return earnings
    except KeyError:
        print(f"KeyError: No earnings dates found for {ticker}. Skipping...")
        return None
    except Exception as e:
        print(f"An error occurred for {ticker}: {e}. Skipping...")
        return None

# URL of the Wikipedia page containing the list of S&P 500 companies
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'

# Read the table from the Wikipedia page
tables = pd.read_html(url)

# The first table on the page is the list of S&P 500 companies
sp500_table = tables[0]

# Extract the ticker symbols into a new DataFrame
tickers_df = sp500_table[['Symbol']]

# Initialize a list to store DataFrames of earnings dates
earnings_dates_list = []

# Loop over each ticker in the DataFrame and get the earnings dates
for ticker in tickers_df['Symbol']:
    earnings_dates = get_earnings_dates(ticker)
    if earnings_dates is not None and not earnings_dates.empty:
        earnings_dates['Ticker'] = ticker
        earnings_dates_list.append(earnings_dates)

# Concatenate all the individual DataFrames into a single DataFrame
if earnings_dates_list:
    earnings_dates_df = pd.concat(earnings_dates_list).reset_index(drop=True)
else:
    earnings_dates_df = pd.DataFrame(columns=['Earnings Date', 'Info', 'Ticker'])

# Rename the index column to 'Earnings Date'
earnings_dates_df.rename(columns={'index': 'Earnings Date'}, inplace=True)

# Display the earnings dates DataFrame
print(earnings_dates_df)

# Save the DataFrame to a CSV file
earnings_dates_df.to_csv('earnings_dates.csv', index=False)
```

```python
import yfinance as yf
from datetime import datetime, timedelta

def has_weekly_options(ticker):
    try:
        # Create a Ticker object
        stock = yf.Ticker(ticker)
        expirations = stock.options
        
        # Convert expiration dates to datetime objects
        expiration_dates = [datetime.strptime(date, '%Y-%m-%d') for date in expirations]
        
        # Check for weekly intervals
        for i in range(len(expiration_dates) - 1):
            if expiration_dates[i + 1] - expiration_dates[i] <= timedelta(days=7):
                return True
        return False
    except Exception as e:
        print(f"An error occurred for {ticker}: {e}. Skipping...")
        return False

# Check for the ticker TSLA
ticker = 'BA'
has_weekly = has_weekly_options(ticker)

if has_weekly:
    print(f"{ticker} has weekly options.")
else:
    print(f"{ticker} does not have weekly options.")
```

```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

def has_weekly_options(ticker):
    """
    Check if a ticker has options with weekly expirations.

    Parameters:
    ticker (str): The stock ticker symbol.

    Returns:
    bool: True if the ticker has weekly options, False otherwise.
    """
    try:
        # Create a Ticker object
        stock = yf.Ticker(ticker)
        
        # Get the expiration dates for options
        expirations = stock.options
        
        # Convert expiration dates to datetime objects
        expiration_dates = [datetime.strptime(date, '%Y-%m-%d') for date in expirations]
        
        # Check for weekly intervals
        for i in range(len(expiration_dates) - 1):
            if expiration_dates[i + 1] - expiration_dates[i] <= timedelta(days=7):
                return True
        return False
    except Exception as e:
        print(f"An error occurred for {ticker}: {e}. Skipping...")
        return False

# URL of the Wikipedia page containing the list of S&P 500 companies
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'

# Read the table from the Wikipedia page
tables = pd.read_html(url)

# The first table on the page is the list of S&P 500 companies
sp500_table = tables[0]

# Extract the ticker symbols into a new DataFrame
tickers_df = sp500_table[['Symbol']]

# Initialize a list to store tickers with weekly options
tickers_with_weekly_options = []

# Loop over each ticker in the DataFrame and check for weekly options
for ticker in tickers_df['Symbol']:
    if has_weekly_options(ticker):
        tickers_with_weekly_options.append(ticker)

# Display the tickers with weekly options
print("Tickers with weekly options:")
print(tickers_with_weekly_options)

# Optionally, save the list to a CSV file
tickers_with_weekly_options_df = pd.DataFrame(tickers_with_weekly_options, columns=['Ticker'])
tickers_with_weekly_options_df.to_csv('tickers_with_weekly_options.csv', index=False)
```
