```python
import pandas as pd
from datetime import datetime
from finvizfinance.screener.overview import Overview

now = datetime.now()

# Format the date and time as a string
date_str = now.strftime("%Y%m%d")
```

```python
# All Biotech companies
foverview = Overview()
filters_dict = {'Sector':'Healthcare', 'Industry':'Biotechnology'}

foverview.set_filter(filters_dict=filters_dict)
df = foverview.screener_view()

print("")
print("Trobades ", len(df), " empreses")
```

```python
# All Biotech companies
foverview = Overview()
filters_dict = {'Sector':'Healthcare', 'Industry':'Biotechnology', 'IPO Date':'In the last 5 years', 'Option/Short':'Optionable'}

foverview.set_filter(filters_dict=filters_dict)
df = foverview.screener_view()

print("")
print("Trobades ", len(df), " empreses")
```

```python
df
```

```python
pd.set_option('display.float_format', lambda x: '{:,.0f}'.format(x))
df = df.sort_values(by='Market Cap', ascending=False)
df = df.reset_index()
df.head(30)
```

```python
output_dir = "output"
filename = f"{output_dir}/Biotech_{date_str}.csv"

# Save DataFrame to CSV
df.to_csv(filename, sep=";", index=False)

print(f"DataFrame saved to {filename}")
```

```python
import pandas as pd
from finvizfinance.insider import Insider

# Create an object for insider trading data
insider_data = Insider()

# Get the latest insider trading data
latest_insider_data = insider_data.get_insider()

# Convert the data to a DataFrame (if it's not already one)
if isinstance(latest_insider_data, list):
    latest_insider_data = pd.DataFrame(latest_insider_data)

# Print the latest insider trading data
for index, row in latest_insider_data.iterrows():
    print(f"Ticker: {row['Ticker']}")
    print(f"Owner: {row['Owner']}")
    print(f"Relationship: {row['Relationship']}")
    print(f"Transaction Date: {row['Date']}")
    print(f"Transaction Type: {row['Transaction']}")
    print(f"Cost: {row['Cost']}")
    print(f"Shares: {row['#Shares']}")
    print(f"Value: {row['Value ($)']}")
    print(f"Shares Total: {row['#Shares Total']}")
    print(f"SEC Form 4 Link: {row['SEC Form 4 Link']}")
    print(f"SEC Form 4: {row['SEC Form 4']}\n")
```

```python
import yfinance as yf
import pandas as pd

# Specify the ticker symbol
ticker_symbol = "BNTX"  # Example: Apple Inc.

# Create a Ticker object
ticker = yf.Ticker(ticker_symbol)

# Fetch historical dividend data
dividends = ticker.dividends

# Check if dividends data is not empty
if not dividends.empty:
    # Convert the dividends Series to a DataFrame
    dividends_df = dividends.reset_index()
    dividends_df.columns = ['Date', 'Dividend']

    # Format the 'Dividend' column to show two decimal places
    dividends_df['Dividend'] = dividends_df['Dividend'].map(lambda x: f"{x:.2f}")
else:
    print("No dividend data available for the specified ticker symbol.")
dividends_df
```
